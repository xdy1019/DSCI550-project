**BAR CHART SHOWING DIFFERENT HOME TYPES**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/transformed_dataset.csv')

# Compute average price for each home type
avg_price = df.groupby('type')['price'].mean().reset_index()

# Sort for a cleaner look
avg_price = avg_price.sort_values('price', ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x='type', y='price', data=avg_price, palette='viridis')
plt.title('Average Price by The Type of Home')
plt.xlabel('Type of Home')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=50)
plt.show()
```

**BAR CHART SHOWCASING HOW MANY HOMES ALLOW FOR WHAT CHARACTERISTICS**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/transformed_dataset.csv')

# Columns to analyze
features = ['cats_allowed', 'dogs_allowed', 'smoking_allowed',
            'wheelchair_access', 'electric_vehicle_charge', 'comes_furnished']

# Count how many listings allow each feature
allow_counts = df[features].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=allow_counts.index, y=allow_counts.values, palette='mako')
plt.title('Number of Homes That Allow Each Characteristic')
plt.xlabel('Types of Characteristics')
plt.ylabel('Count of Homes')
plt.xticks(rotation=50)
plt.show()
```

**BAR CHART SHOWING BOTH YES AND NO FOR CHARACTERISTICS**

```python
avg_prices = {}
for col in features:
    avg = df.groupby(col)['price'].mean()
    avg_prices[col] = avg

# Convert to DataFrame for easy plotting
avg_df = pd.DataFrame(avg_prices).T
avg_df.columns = ['Not Allowed', 'Allowed']

# Plot side-by-side bars
avg_df.plot(kind='bar', figsize=(9,5))
plt.title('Average Price by Characteristic Allowance')
plt.ylabel('Average Price ($)')
plt.xlabel('Characteristic (Allowance)')
plt.xticks(rotation=50)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

**SCATTER PLOT FOR AVG PRICE VS AVG SQUARE FEET**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/transformed_dataset.csv')

# Compute average price and average square footage per home type
avg_data = df.groupby('type')[['price', 'sqfeet']].mean().reset_index()

# Sort by average price (for readability)
avg_data = avg_data.sort_values('price', ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='sqfeet', y='price', data=avg_data, hue='type', s=100)
plt.title('Average Price vs. Average Square Feet by Home')
plt.xlabel('Square Feet')
plt.ylabel('Price ($)')
plt.legend(title='Types of Homes')
plt.show()
```

**BAR CHART SHOWCASING AVG PRICE BASED ON REGION AND TYPE OF HOME**

```python
plt.figure(figsize=(10,6))
sns.barplot(x='region', y='price', hue='type', data=df, ci=None)
plt.title('Average Price Based on Region and Type of Home')
plt.xlabel('Region')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Type of Home')
plt.tight_layout()
plt.show()
```

**BAR CHART SHOWCASING AVG PRICE BY REGION**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/transformed_dataset.csv')

# Compute average price per region
avg_price_region = df.groupby('region')['price'].mean().reset_index()

# Sort by price (descending)
avg_price_region = avg_price_region.sort_values('price', ascending=False)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x='region', y='price', data=avg_price_region, palette='coolwarm')
plt.title('Average Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**BAR CHART SHOWCASING NUMBER OF HOMES PER REGION**

```python
region_counts = df['region'].value_counts().reset_index()
region_counts.columns = ['Region', 'Number of Homes']

plt.figure(figsize=(10,5))
sns.barplot(x='Region', y='Number of Homes', data=region_counts, palette='viridis')
plt.title('Number of Homes by Region')
plt.xlabel('Region')
plt.ylabel('Count of Homes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
